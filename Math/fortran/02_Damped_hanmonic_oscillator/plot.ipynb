```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
DP_df = pd.read_csv('./X_V.csv')

plt.figure(figsize=(20, 4))
plt.subplot(1, 3, 1)
plt.plot(DP_df[' Time'],DP_df['X'])
plt.xlabel('Time [s]')
plt.ylabel('Position [m]')
plt.legend(['Forward Euler', 'ODE45 (RK4)'])
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(DP_df[' Time'],DP_df['V'])
plt.xlabel('Time [s]')
plt.ylabel('Velocity [m/s]')
plt.legend(['Forward Euler', 'ODE45 (RK4)'])
plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(DP_df['X'],DP_df['V'])
plt.xlabel('Position [m]')
plt.ylabel('Velocity [m/s]')
plt.legend(['Forward Euler', 'ODE45 (RK4)'])
plt.grid(True)
```
